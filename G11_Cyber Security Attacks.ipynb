**Análisis temporal de ciberataques:**
¿Hay patrones de tiempo específicos (diurnos, nocturnos, días de la semana) en los ataques según las marcas de tiempo?


```python
# Convertir la columna de Timestamp a datetime
# Se cambió 'ts' por 'Timestamp' que es el nombre correcto de la columna en la DB
df['timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')

# Extraer información temporal relevante
df['hour'] = df['timestamp'].dt.hour  # Extraer la hora
df['day_of_week'] = df['timestamp'].dt.dayofweek  # Extraer el día de la semana (0 = lunes, 6 = domingo)

# Clasificar ataques como diurnos o nocturnos
def clasificar_horario(hora):
    if 6 <= hora < 18:
        return 'diurno'
    else:
        return 'nocturno'

df['time_of_day'] = df['hour'].apply(clasificar_horario)

# Agrupar ataques por franja horaria (diurno/nocturno)
ataques_por_franja = df['time_of_day'].value_counts()

# Agrupar ataques por día de la semana
ataques_por_dia = df['day_of_week'].value_counts().sort_index()

# Mostrar resultados
print("Distribución de ataques por franja horaria:")
print(ataques_por_franja)

print("\nDistribución de ataques por día de la semana (0=Lunes, 6=Domingo):")
print(ataques_por_dia)
```

El análisis muestra que la mayoría de los ciberataques ocurren durante la noche con 40mil ataques. La distribución por días es constante, con ligeros picos los domingos, resaltando la necesidad de vigilancia continua y estrategias de seguridad en horarios críticos.

```python
# Primero filtraremos solo los ataques con alto nivel de severidad.
# Supongamos que los niveles de severidad son 'High', 'Medium', 'Low'
high_severity_attacks = df[df['Severity_Level'] == 'high'] # Changed 'data' to 'df' and 'Severity Level' to 'Severity_Level' to match the column name after cleaning

# Mostrar las primeras filas de los ataques de alta severidad
print(high_severity_attacks.head())

# Posterior a eso contaremos los protocolos más comunes en ataques de alta severidad.
protocol_counts = high_severity_attacks['Protocol'].value_counts().reset_index()
protocol_counts.columns = ['Protocol', 'Count']

# Por ultimo mostraremos los protocolos más frecuentes asociados a los ataques de mayor severidad.
print(protocol_counts.head(10))
```

**Identificación de puertos vulnerables:**
¿Cuáles son los puertos de destino más comúnmente atacados y qué tipo de tráfico se dirige a ellos?


```python
# Agrupar por Destination_Port y Traffic_Type para contar las ocurrencias
puertos_trafico = df.groupby(['Destination_Port', 'Traffic_Type']).size().reset_index(name='count')

# Ordenar los resultados por la cantidad de ocurrencias en orden descendente
puertos_trafico = puertos_trafico.sort_values(by='count', ascending=False)

# Mostrar los resultados
print(puertos_trafico)

# Agrupar solo por Destination_Port y contar las ocurrencias
puertos_mas_atacados = df['Destination_Port'].value_counts().reset_index()
puertos_mas_atacados.columns = ['Destination_Port', 'count']

# Mostrar los 10 puertos más atacados
print("Puertos más comúnmente atacados:")
print(puertos_mas_atacados.head(10))

# Filtrar los 10 puertos más atacados
top_puertos = puertos_mas_atacados.head(10)['Destination_Port']

# Filtrar el DataFrame original para los puertos más atacados
df_top_puertos = df[df['Destination_Port'].isin(top_puertos)]

# Agrupar por Destination_Port y Traffic_Type
trafico_por_puerto = df_top_puertos.groupby(['Destination_Port', 'Traffic_Type']).size().reset_index(name='count')

# Ordenar para visualización
trafico_por_puerto = trafico_por_puerto.sort_values(by=['Destination_Port', 'count'], ascending=[True, False])

# Mostrar los resultados
print("Tipo de tráfico dirigido a los puertos más atacados:")
print(trafico_por_puerto)

```

Los puertos **34117** y **7508** son los más atacados.
El tráfico HTTP domina, seguido por DNS y FTP, lo que sugiere que los atacantes buscan vulnerabilidades en servicios web, resolución de nombres y transferencias de archivos.
La diversidad de tráfico en los puertos más atacados indica que los atacantes intentan múltiples enfoques para comprometer los servicios.

**Relación entre firmas de ataque y acciones tomadas:**
¿Qué firmas de ataque tienen una mayor probabilidad de desencadenar acciones específicas (como bloqueos, alertas, etc.)?



```python
# Agrupamos Attack_Signature y Action_Taken para contar cuántas veces ocurre cada combinación.

# Contar las ocurrencias de cada combinación
resumen = df.groupby(['Attack_Signature', 'Action_Taken']).size().reset_index(name='count')

# Calcular la proporción relativa por cada firma de ataque
resumen['proporcion'] = resumen['count'] / resumen.groupby('Attack_Signature')['count'].transform('sum')

# Convertir la proporción en porcentaje
resumen['proporcion_%'] = resumen['proporcion'] * 100

print(resumen)
```

Para ambas firmas de ataque (known pattern a y known pattern b), las acciones tomadas (blocked, ignored, y logged) tienen proporciones muy similares.
Esto sugiere que no hay un sesgo evidente hacia una acción en particular para estas firmas de ataque. En otras palabras, el sistema reacciona de manera equilibrada ante estas firmas.
